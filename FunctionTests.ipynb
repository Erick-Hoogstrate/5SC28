```python
from matplotlib import pyplot as plt
from data import load_data, load_narx_data
```

```python
x_train, y_train = load_data(section="train", split=[0.6, 0.2, 0.2], total_number_of_points=2000, as_tensor=True)

plt.figure(figsize=(12,5)) 
plt.plot(x_train, y_train, '.') 
plt.title('Training data')
plt.grid(); plt.xlabel('x'); plt.ylabel('y');
plt.show() 
```
