```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"disc-benchmark-files\training-data.csv").to_numpy()

sampling_time = 0.025 # 25 ms sampling time is used for obtaining the training data. See discussion forum

x = np.arange(len(data)) * sampling_time/60


plt.figure(figsize=(12,5))
plt.plot(x, data[:,0], '.');
plt.title('Training data - Input')
plt.grid(); plt.xlabel('minutes'); plt.ylabel('Input voltage u');
plt.show() 

plt.figure(figsize=(12,5))
plt.plot(x, data[:,1], '.');
plt.title('Training data - Output')
plt.grid(); plt.xlabel('minutes'); plt.ylabel('Output angle theta');
plt.show() 
```
